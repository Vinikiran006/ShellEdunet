```python
print("Sanity Check of Column Names:")
for col in df.columns:
    print(f"{col} -> Unique: {df[col].nunique()}, Missing: {df[col].isna().sum()}")
```

## sanity gives if there are any repeated values or if they are unique it returns the number of unique and missing values

```python
for col in num_cols:
    z_scores = (df[col] - df[col].mean())/df[col].std()
    outliers = (np.abs(z_scores) > 3).sum()
    print(f"{col} -> Outliers: {outliers}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# List of categorical columns
cat_cols = ["Soil Type", "Crop Type", "Fertilizer"]

# Plot bar plots for each categorical column
for col in cat_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, data=df)
    plt.title(f"Bar Plot of {col}")
    plt.xticks(rotation=45)  # rotate labels if long
    plt.show()
```

```python
# Example: Average Temperature by Crop Type
sns.barplot(x="Crop Type", y="Temperature", data=df)
plt.title("Average Temperature per Crop Type")
plt.xticks(rotation=45)
plt.show()

# Example: Average Nitrogen by Soil Type
sns.barplot(x="Soil Type", y="Nitrogen", data=df)
plt.title("Average Nitrogen per Soil Type")
plt.xticks(rotation=45)
plt.show()
```
